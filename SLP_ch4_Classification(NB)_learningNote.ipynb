````python
from keras.models import Sequential
from keras.layers import Dense
from matplotlib import pyplot
````

##### define model



```
# Dense input_shape:
#    N-D tensor with shape: (batch size, ..., input dim)
#    most common input with shape -- 2D : (batch size, input dim)
#    don't need to specify after the first layer
```



````python
modelK = Sequential()
modelK.add(Dense(100, input_shape=(22,), activation='elu')) # input shape == feature column#
modelK.add(Dense(64, activation='relu'))
modelK.add(Dense(1, activation='sigmoid'))
````

##### compile model

[CSDN](https://blog.csdn.net/yunfeather/article/details/106461754), [TensorFlow](https://www.tensorflow.org/api_docs/python/tf/keras/Model)
<br> The compile method is used to configure the learning process, which takes three main parameters:

*  [`loss:`](https://www.tensorflow.org/api_docs/python/tf/keras/losses) The loss function to minimize during training. For example, 'categorical_crossentropy' for multi-class classification.
*  [`optimizer:`](https://www.tensorflow.org/api_docs/python/tf/keras/optimizers) The optimizer algorithm such as 'sgd' for stochastic gradient descent.
*  `metrics:` The list of metrics to monitor during training and testing.

````python
modelK.compile(loss='binary_crossentropy', optimizer='adagrad', metrics=['accuracy'])
````

##### fit model

````python
history = modelK.fit(feature_train, label_train,
                    validation_data=(feature_test, label_test),
                    epochs=80, verbose=1)
````

## Bootstraping

````python
import random
import matplotlib.pyplot as plt
from scipy.stats import bootstrap

f1_scores_train = []
f1_scores_test = []
avg_f1 = dict()
avg_f1['train'] = []
avg_f1['test'] = []
for i in range(300):
  subset200 = random.sample(range(0, len(label_test)) , 200)
  unique_numbers = len(set(subset200)) == 200
  ftest = feature_test.iloc[subset200]
  g_ltest = label_test.iloc[subset200]
  p_ltest = multiNB.predict(ftest)
  f1_scores_test.append(f1_score(g_ltest, p_ltest, average='micro'))
  avg_f1['test'].append(sum(f1_scores_test) / len(f1_scores_test))
  ftrain = feature_train.iloc[subset200]
  g_ltrain = label_train.iloc[subset200]
  p_ltrain = multiNB.predict(ftrain)
  f1_scores_train.append(f1_score(g_ltrain, p_ltrain, average='micro'))
  avg_f1['train'].append(sum(f1_scores_train) / len(f1_scores_train))
````

#### Plotting

````python
plt.subplots()
plt.plot(f1_scores_train, label = "Train")
plt.plot(f1_scores_test, label = "Test")
plt.title("F1-score during 300 times bootstraping")
plt.xlabel("Bootstrap Iteration")
plt.ylabel("F1-score")
plt.legend()
plt.show()

plt.subplots()
plt.plot(avg_f1['train'], label = "Train")
plt.plot(avg_f1['test'], label = "Test")
plt.title("F1-score during 300 times bootstraping")
plt.xlabel("Bootstrap Iteration")
plt.ylabel("Average F1-score")
plt.legend()
plt.show()

plt.subplots()
plt.ylim(0,1)
plt.plot(avg_f1['train'], label = "Train")
plt.plot(avg_f1['test'], label = "Test")
plt.title("F1-score during 300 times bootstraping")
plt.xlabel("Bootstrap Iteration")
plt.ylabel("Average F1-score")
pyplot.legend()
plt.show()
````

Search for the best parameter

````python
from sklearn.model_selection import GridSearchCV
def optimize_random_forest_parameters(X, y):
    """
    Optimize random forest parameters using GridSearchCV.

    Parameters:
    X (pd.DataFrame): The feature set.
    y (pd.Series): The target variable.

    Returns:
    dict: The best parameters found by GridSearchCV.
    GridSearchCV: The GridSearchCV object (fitted).
    """
    # Define the parameter grid
    param_grid = {
        'n_estimators': [10, 200, 300, 500, 1000],
        'max_depth': [None, 10, 20, 30],
        'min_samples_split': [2, 5, 10],
        'min_samples_leaf': [1, 2, 4, 10],
        'bootstrap': [True, False],
        'warm_start': [True, False]
    }

    # Create a RandomForestClassifier object
    rf = RandomForestClassifier(random_state=42)

    # Instantiate the GridSearchCV object
    grid_search = GridSearchCV(estimator=rf, param_grid=param_grid, cv=5, n_jobs=-1, verbose=2)

    # Fit the GridSearchCV object to the data
    grid_search.fit(X, y)

    # Get the best parameters and the fitted GridSearchCV object
    best_params = grid_search.best_params_
    return best_params, grid_search
bs1, bp1 = optimize_random_forest_parameters(feature_train, label_train)
````
